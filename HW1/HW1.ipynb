1. To show that T is invertible we have to show that the determinant of matrix T is not zero. If the determinent of matrix T is zero, it indicates that the transformation T collapses one or more spaces of the orignal vector or matrix therefore it is impossible to revert the transformation. To find the determinant of T we can simply call the np.linalg.det function on T.

```python
np.linalg.det(T)
```

Since the determinant of T is not zero, this indicates that it is possible to invert T.

2. To compute T inverse we can call the np.linalg.inv function on T.

```python
Tinv = np.linalg.inv(T)
```
